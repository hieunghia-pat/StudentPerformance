## Two-way ANOVA for family_atts_data

```python
family_atts_data = data[["address", "Pstatus", "Medu", "Mjob", "Fedu", "Fjob", "guardian", "famsize", "famrel", "G3"]]
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

## Two-way ANOVA for all attributes

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
anova_results = []
for att_idx_1 in range(len(data.columns[:-3])):
    for att_idx_2 in range(att_idx_1+1, len(data.columns[:-3])):
        formula = f"G3 ~ C({data.columns[att_idx_1]}) + C({data.columns[att_idx_2]}) + C({data.columns[att_idx_1]}):C({data.columns[att_idx_2]})"
        model = ols(formula, data=data).fit()
        anova_result = sm.stats.anova_lm(model, typ=2)
        anova_result.loc[:, "att1"] = data.columns[att_idx_1]
        anova_result.loc[:, "att2"] = data.columns[att_idx_2]
        anova_results.append(anova_result)
```

```python
anova_results[0]
```

```python
effect_atts = []
for result in anova_results:
    if result["PR(>F)"][2] <= 0.05:
        effect_atts.append({
            "attribute_1": result["att1"][2],
            "attribute_2": result["att2"][2],
            "P": result["PR(>F)"][2]
        })
```

```python
effect_atts
```

```python
len(effect_atts)
```

```python
import json

json.dump(effect_atts, open("effect_attr.json", "w+"))
```

## One-way ANOVA for all attributes

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
anova_results = []
for att_idx in range(len(data.columns[:-3])):
        formula = f"G3 ~ C({data.columns[att_idx]})"
        model = ols(formula, data=data).fit()
        anova_result = sm.stats.anova_lm(model, typ=2)
        anova_result.loc[:, "att"] = data.columns[att_idx]
        anova_results.append(anova_result)
```

```python
anova_results[0]
```

```python
effect_atts = []
for result in anova_results:
    if result["PR(>F)"][0] <= 0.05:
        effect_atts.append({
            "attribute": result["att"][0],
            "P": result["PR(>F)"][0]
        })
```

```python
import json

json.dump(effect_atts, open("effect_single_attr.json", "w+"))
```
